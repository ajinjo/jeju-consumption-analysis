```python
import folium

df_sample['위치좌표 X축값 '].mean()
df_sample['위치좌표 Y축값 '].mean()
```

```python
map=folium.Map(location=[df_sample['위치좌표 Y축값 '].mean(),df_sample['위치좌표 X축값 '].mean()],zoom_start=10)

map
```

```python
map=folium.Map(location=[df_sample['위치좌표 Y축값 '].mean(),df_sample['위치좌표 X축값 '].mean()],zoom_start=10)


for n in df_sample.index:
    name=df_sample.loc[n,'구분']
    
    location=[df_sample.loc[n,'위치좌표 Y축값 '], df_sample.loc[n,'위치좌표 X축값 ']]
    adress=df_sample.loc[n,'소재지']
    popup=f'{name}-{adress}'

    folium.Marker(
        location = location,
        popup = popup,
    ).add_to(map)
map
```
